##Step 6: Data Visualization

```python
# Assigning data to df variable
df = pd.DataFrame(data)

# Creating a pie chart for PRI_FDA Industry Name
df['PRI_FDA Industry Name'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('PRI_FDA Industry Name',weight='bold',fontsize=14)
plt.legend(loc='upper right', fontsize=12, shadow=True,bbox_to_anchor=(2,1))  # Set legend properties
plt.show()
```

```python
# Creating pie chart for CI_Gender
colors = ['gold', 'yellowgreen', 'lightcoral']
df['CI_Gender'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=colors,startangle=90)
plt.title('CI_Gender',weight='bold',fontsize=14)
plt.legend(loc='upper right', fontsize=12, shadow=True,bbox_to_anchor=(2,1))
plt.xticks(rotation=140)
plt.show()
```

```python
# Creating bar chart for top 10 CI_Age Unit
plt.figure(figsize=(6,4))
df['CI_Age Unit'].value_counts().head(10).plot(kind='bar',color='#FF0000',fontsize=10,edgecolor='black',width=0.8)
plt.xlabel('CI_Age Unit',weight='bold',fontsize=14)
plt.ylabel('Count',weight='bold',fontsize=14)
plt.title('CI_Age Unit',weight='bold',fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

```python
# Creating  bar chart of CI_Age at Adverse Event
plt.figure(figsize=(6,4))
df['CI_Age at Adverse Event'].value_counts().head(10).plot(kind='bar',color='#1f77b4',edgecolor='black',width=0.8)
plt.xlabel('CI_Age at Adverse Event')
plt.ylabel('Count')
plt.title('CI_Age at Adverse Event')
plt.xticks(rotation=90)
plt.show()
```

```python
# Creating histogram of PRI_FDA Industry Name
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.hist(df['PRI_FDA Industry Name'],color='#ff7f0e',edgecolor='black')
plt.title('PRI_FDA Industry Name')
plt.xlabel('PRI_FDA Industry Name')
plt.ylabel('Frequency')

# Creating histogram of CI_Age Unit at Adverse Event
plt.subplot(2, 2, 2)
plt.hist(df['CI_Age Unit'], color='#ff7f0e',edgecolor='black')
plt.title('CI_Age Unit')
plt.xlabel('CI_Age Unit ')
plt.ylabel('Frequency')



```
